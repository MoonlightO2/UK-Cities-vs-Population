```python
# Initialize the Folium map
m = folium.Map(location=[54.7023545, -3.2765753], zoom_start=6)

# Iterate over rows in the DataFrame and add markers to the map
for index, row in data.iterrows():
    country = row['Country']
    city = row['City']
    year = row['Year granted or confirmed']
    region = row['Region']
    nation = row['Nation']
    population = row['Population']

    # Skip row if 'Year granted or confirmed' is not a valid integer
    try:
        int_year = int(year)
        century = (int_year // 100) + 1 if int_year % 100 != 0 else int_year // 100
    except ValueError:
        # If 'Year granted or confirmed' is not a valid integer, set century to "Time immemorial"
        century = "Time immemorial"

    # Get coordinates using the get_coordinates function
    coordinates = get_coordinates(api_key, city, country)

    # Define a color map for pastel colors based on centuries
    color_map = {
        '10': '#FFD700',  # Gold
        '11': '#FF69B4',  # Hot Pink
        '12': '#87CEEB',  # Sky Blue
        '13': '#FFA07A',  # Light Salmon
        '14': '#20B2AA',  # Light Sea Green
        '15': '#FF6347',  # Tomato
        '16': '#FFD700',  # Gold (repeated for demonstration)
        '17': '#98FB98',  # Pale Green
        '18': '#FFB6C1',  # Light Pink
        '19': '#9370DB',  # Medium Purple
        '20': '#FFA500',  # Orange
        '21': '#00FA9A',  # Medium Spring Green
        'Time immemorial': '#FF0000'  # Red (repeated for demonstration)
    }

    # Get coordinates using the get_coordinates function
    coordinates = get_coordinates(api_key, city, country)

    # Add marker to the map with a smaller glowing circle icon and pastel color
    if coordinates:
        popup_content = f"Country: {country}\n\nCity: {city}\n\nCentury: {century}\n\nRegion: {region}\n\nNation: {nation}\n\nPopulation: {population}"
        color = color_map.get(str(century), '#FFD700')  # Default to gold color for unspecified centuries

        # Check for "Time immemorial" and use red color
        if century == "Time immemorial":
            color = color_map['Time immemorial']

        folium.CircleMarker(
            location=[coordinates[0], coordinates[1]],
            radius=5,
            color=color,
            fill=True,
            fill_color=color,
            popup=folium.Popup(popup_content, parse_html=True, max_width='500px')  # Set max_width to the desired value
        ).add_to(m)

# Save the map to an HTML file
m.save(map_path)
display(m)
```
